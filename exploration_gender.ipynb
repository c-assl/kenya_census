```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as plt
```

```python
gender = pd.read_csv("data/gender.csv")
gender
```

##### pistes d'explo : rassembler les county en regions plus grandes ??
https://en.wikipedia.org/wiki/Counties_of_Kenya : intéressant avoir carte

```python
gender_melt = gender.drop(labels = "Total", axis = 1).drop(index = 0).melt(id_vars=["County"], var_name="gender", value_name="number")
gender_melt
```

```python
# sns.distplot(
#     x = ,
#     y = ,
#     data = gender_melt
# )

sns.countplot(
      x="number",
      hue=gender_melt.gender, # sensé $etre remplaçable par col
      data=gender_melt
)
```

```python
sns.set(rc = {'figure.figsize':(30,15)})

sns.barplot(
    x = "County",
    y = "number",
    hue = "gender",
    data = gender_melt
)
```

```python
plt.barplot(
    x = "County",
    y = "number",
    hue = "gender",
    data = gender_melt
)
```
